```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data
n = 2 # features
m = 100 # samples

data = np.random.random(size=(m,n)) # (m, n)
data.shape

# nubmer of cluster 
k = 3
n_iter = 5
distortion_history = []
# random centroids
np.random.seed(1)
centroids = np.random.choice(data.shape[0], size=k, replace=False)
centroids = data[centroids] # (k,n)

for epoch in range(n_iter):
    # distance to the centroids
    diff = data[:, np.newaxis, :] - centroids # (m, k, n)
    # norms of the distances sqrt of sum of square the elements in an array
    norms = np.linalg.norm(diff, axis=2) # (m, k)
    distortion_history.append(np.sum(np.min(norms, axis=1))) # scalar
    # print(norms, norms.shape)
    clusters = np.argmin(norms, axis=1) # (m,)
    print(clusters)

    # new centroids
    for cluster in range(k):
        centroids[cluster] = np.mean(data[clusters == cluster], axis= 0)

    print("new centroids:\n", centroids)
 
print("distortion: \n", distortion_history)
```

```python
plt.figure(figsize=(10,5))
plt.plot(distortion_history)
plt.show()
```
